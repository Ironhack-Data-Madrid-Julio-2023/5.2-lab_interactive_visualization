## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
df = pd.read_excel('../data/Online Retail.xlsx')

df.head()
```

```python
df[(df.Country != 'United Kingdom')&(df.InvoiceDate <= '2011-04-30')&(df.InvoiceDate >= '2011-04-01')]\
.groupby('Country').sum()[['Quantity','Revenue']]\
.iplot(kind = 'bar', xTitle = 'Country', title = 'Cantidad total y beneficio');
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
dfy = df.copy()

dfy = dfy[(dfy.InvoiceDate >= '2011-01-01')&(dfy.InvoiceDate <= '2011-05-31')&(dfy.Country == 'France')]\
[['Quantity','Revenue','InvoiceDate']].groupby('InvoiceDate').sum()\
.iplot(kind = 'line', xTitle = 'Invoice Date', title = 'Cantidad total y beneficio a lo largo del tiempo');
# Usamos sum() porque la agrupacin simplemente nos sirve para convertir InvoiceDate en el ndice, ya que son
# valores nicos.
```

## 3. Create an interactive scatter plot showing the relationship between average quantity (x-axis) and average unit price (y-axis) for the product PARTY BUNTING with the plot points color-coded by country (categories).

```python
dff = df.copy()
dff = dff[dff.Description == 'PARTY BUNTING'].groupby('Country').mean()[['Quantity', 'UnitPrice']]

dff.reset_index(inplace=True)

dff.iplot(x = 'Quantity', 
       y = 'UnitPrice', 
       categories = 'Country',
       xTitle = 'Average quantity', 
       yTitle = 'Average Unit Price',
       color = 'blue')
```

## 4. Create a set of interactive histograms showing the distributions of quantity per invoice for the following countries: EIRE, Germany, France, and Netherlands.

```python
layout = go.Layout(
        xaxis=dict(
            range=[0, 400]
        )
    )

df[df.Country == 'EIRE']['Quantity'].iplot(kind = 'hist', layout = layout) 
# Ejes reducidos para ver mejor la distribucin, ignorando outliers
```

```python
layout = go.Layout(
        xaxis=dict(
            range=[0, 150]
        )
    )

df[df.Country == 'Germany']['Quantity'].iplot(kind = 'hist', layout = layout)
```

```python
layout = go.Layout(
        xaxis=dict(
            range=[0, 150]
        )
    )

df[df.Country == 'France']['Quantity'].iplot(kind = 'hist', layout = layout)
```

```python
layout = go.Layout(
        xaxis=dict(
            range=[0, 550]
        )
    )

df[df.Country == 'Netherlands']['Quantity'].iplot(kind = 'hist', layout = layout)
```
